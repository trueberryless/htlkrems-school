```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loss_function_f(x):
    return x**2 - 4 * x +5
```

```python
def derivation_f(x):
    return 2*x - 4
```

```python
# Step 1 - Funktion ausgeben!
xs = np.arange(-2, 6, 0.1)
ys = loss_function_f(xs)
plt.plot(xs,ys)
plt.show()
```

```python
# Step 2 - Startpunkt hinzufügen!
xs = np.arange(-2, 6, 0.1)
ys = loss_function_f(xs)

x = 5


plt.scatter(x, loss_function_f(x), c="r")
plt.plot(xs,ys)
plt.show()
```

```python
# Step 3 - Jetzt wandern wir nach unten, indem wir einen Punkt hinzufügen!
xs = np.arange(-2, 6, 0.1)
ys = loss_function_f(xs)

x = 5
plt.scatter(x, loss_function_f(x), c="r")
steigung = derivation_f(x)
print(steigung)
x = x - steigung
# Neues x ist jetzt ein bisschen zu weit gegangen!
plt.scatter(x, loss_function_f(x), c="g")
plt.plot(xs,ys)
plt.show()
```

```python
# Step 4 - Jetzt mit Lernrate!
xs = np.arange(-2, 6, 0.1)
ys = loss_function_f(xs)

x = 5
plt.scatter(x, loss_function_f(x), c="r")
steigung = derivation_f(x)
print(steigung)
x = x - 0.01*steigung
# Jetzt ein wenig besser - zumindes in die richtige Richtung und nicht zu weit!
plt.scatter(x, loss_function_f(x), c="r")
plt.plot(xs,ys)
plt.show()
```

```python
# Step 5 - Jetzt mit Iteration!
xs = np.arange(-2, 6, 0.1)
ys = loss_function_f(xs)

x = 5

for steps in range(20):
    steigung = derivation_f(x)
    x = x - 0.2 * steigung
    #print(loss_function_f(x))
    plt.scatter(x, loss_function_f(x), c="r")

print("Steigung:", steigung)
print("x", x)
plt.scatter(x, loss_function_f(x), c="r")
plt.plot(xs,ys)
plt.show()
#print(ys)
```

```python
Y_true = [1,1,2,2,4]

Y_pred = [0.6, 1.2, 1.5, 2.6, 3]

# Mean Squared Error
mse = np.square(np.subtract(Y_true,Y_pred)).mean()
print("MSE: ", mse)

# Step by step:
diffs = np.subtract(Y_true,Y_pred)
print(diffs)

squared_diffs = np.square(np.subtract(Y_true,Y_pred))
print(squared_diffs)
print(squared_diffs.sum())

print("MSE: ", squared_diffs.mean())
```
